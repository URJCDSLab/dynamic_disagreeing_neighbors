```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from src.model.ddn import *
from src.model.instance_hardness import *
from src.model.ddn import *
from sklearn.model_selection import train_test_split

np.random.seed(1234)

n = 50

mu1, mu2, sigma = 0, -2, 1

X1 = np.random.normal(mu1, sigma, size=[n, 2])
X2 = np.random.normal(mu2, sigma, size=[n, 2])
X3 = np.random.normal([-2, 2], 0.5, size=[n, 2])
X = np.vstack((X1, X2, X3))

y = np.hstack((np.zeros(int(len(X1))),np.ones(int(len(X2) + len(X3))))).astype(int)
```

```python
ddn_complexity = DDN(k=3)
ddn_complexity.fit(X, y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Asumimos que tienes X, y, ddn_complexity definidos

figure, axes = plt.subplots(figsize=[8, 8])

# Scatter plot con colores y marcadores segn y
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='tab:blue', marker='o', alpha=0.7)
plt.scatter(X[y == 1, 0], X[y == 1, 1], c='tab:orange', marker='^', alpha=0.7)

for i in [4, 18, 19, 26, 89, 109]:
    neighbours = X[ddn_complexity.support_neighbours[i]]
    target = X[i]
    radius = ddn_complexity.radius[i]

    if i == 4:
        color = 'darkgoldenrod'
        linewidth = 2
    elif i == 26:
        color = 'darkgoldenrod'
        linewidth = 2
    else:
        color = 'black'
        linewidth = 1

    Drawing_uncolored_circle = plt.Circle(target, radius, fill=False, color=color, linewidth=linewidth)

    plt.scatter(target[0], target[1], alpha=0)
    plt.scatter(neighbours[:, 0], neighbours[:, 1], alpha=0.7, s=60, marker='x', color='red')
    axes.set_aspect(1)
    axes.add_artist(Drawing_uncolored_circle)

plt.show()
```
